```python
x = df1
```

```python
from sklearn.model_selection import train_test_split
```

```python
y = df['Admission Chance']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.21)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression(fit_intercept = True)
reg.fit(x_train,y_train)
```

```python
print('Linear Model Coeff (m) ', reg.coef_)
print('Linear Model Coeff (b) ', reg.intercept_)
```

```python
y_predict = reg.predict(x_test)   
y_predict
```

```python
plt.scatter(y_test, y_predict, color ='r')
plt.ylabel('Model Predictions')
plt.xlabel ('True  (Ground Truth)  ')
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
```

```python
k = x_test.shape[1]
n = len(x_test)
```

Evaule the model

```python
RMSE = float(format(np.sqrt(mean_squared_error(y_test,y_predict)) , '.3f'))
MSE = mean_squared_error(y_test,y_predict)
MAE = mean_absolute_error(y_test,y_predict)
r2 = r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2)*(n-1)/(n-k-1)
MAPE = np.mean(np.abs((y_test-y_predict)/y_test))*100
print('RMSE=',RMSE, '\nMSE = ', MSE , '\nMAE=', MAE , '\nr2 =', r2, '\nadj_r2 = ', adj_r2, '\nMAPE=', MAPE)
```
